```python
#Tugas 1
#Dataframe sederhana dan operasi dasar
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("HandsOnWeek3").getOrCreate()

data = [("Bond", "Agen", 3000),
         ("Jackson", "Penyanyi", 4600),
         ("Stark", "Ilmuan", 4100),
         ("Ozawa", "Aktris", 3000)]
columns = ["Nama", "Pekerjaan", "Gaji"]

df = spark.createDataFrame(data, schema=columns)
df.show()
```

```python
#Tugas 2
#Menggunakan operasi sederhana
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("HandsOnWeek 3").getOrCreate()

data = [("Srimul", "Menteri", 3200),
        ("Luhut", "CEO", 2100),
        ("Sahroni", "Maid", 5400),
        ("UyaKuya", "Pesulap", 1000)
        ]
columns = ("Nama", "Pekerjaan", "Gaji")

df = spark.createDataFrame(data, schema=columns)

#Contoh operasi dasar
df.select("Nama", "gaji").show()
df.filter(df["Gaji"] > 2100).show()
df.groupBy("Pekerjaan").min("Gaji").show()

```

```python
#Tugas 3
#Mengeksplorasi mengoolah tipe data kompleks menggunakan Spark Dataframes
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

spark = SparkSession.builder.appName("HandsOnWeek 3").getOrCreate()

data = [("Srimul", "Menteri", 3200),
        ("Luhut", "CEO", 2100),
        ("Sahroni", "Maid", 5400),
        ("UyaKuya", "Pesulap", 1000)]
columns = ("Nama", "Pekerjaan", "Gaji")

df = spark.createDataFrame(data, schema=columns)

#Manipulasi tipe data kompleks
df_bonus = df.withColumn("Bonus Gaji", col("gaji") * 0.1)
df_final = df_bonus.withColumn("Total Kompensasi", col("Gaji") + col("Bonus Gaji"))

df_final.show()
```

```python
#Tugas 4
#Operasi data lanjutan
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

spark = SparkSession.builder.appName("HandsOnWeek 3").getOrCreate()

data = [("Srimul", "Menteri", 3200),
        ("Luhut", "CEO", 2100),
        ("Sahroni", "Maid", 5400),
        ("UyaKuya", "Pesulap", 1000)]
columns = ("Nama", "Pekerjaan", "Gaji")

df = spark.createDataFrame(data, schema=columns)
# Contoh menggunakan window functions
from pyspark.sql.window import Window
from pyspark.sql import functions as F

windowSpec = Window.partitionBy('Pekerjaan').orderBy('Gaji')
df.withColumn('Rank', F.rank().over(windowSpec)).show()
```

```python
#Tugas 5
#Memanipulasi Dataset

# Upload file specific
from google.colab import files
uploaded = files.upload()

# Ekstrak file
import zipfile
with zipfile.ZipFile('archive (1).zip', 'r') as zip_ref:
    zip_ref.extractall("lung_cancer_data")

# Setup Spark
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
spark = SparkSession.builder.appName("LungCancer").getOrCreate()

# Baca dataset
df = spark.read.csv("lung_cancer_data/lung_cancer_dataset.csv",
                   header=True,
                   inferSchema=True)

# Tampilkan data
print("Dataset Shape:", (df.count(), len(df.columns)))
df.show(10)
df.printSchema()

# ==============================================
# MANIPULASI DATA SEDERHANA
# ==============================================

print("\n" + "="*50)
print("MANIPULASI DATA")
print("="*50)

#Filter data - pasien di atas usia 50 tahun
print("Pasien di atas usia 50 tahun:")
df.filter(F.col("age") > 50).show(10)

#Seleksi kolom tertentu
print("Data pasien (ID, usia, gender, kanker):")
df.select("patient_id", "age", "gender", "lung_cancer").show(10)

#Group by dan aggregasi - rata-rata usia per gender
print("Rata-rata usia berdasarkan gender:")
df.groupBy("gender").agg(F.avg("age").alias("rata_rata_usia")).show()

#Urutkan data berdasarkan usia (descending)
print("Data diurutkan berdasarkan usia (tertua ke termuda):")
df.orderBy(F.col("age").desc()).select("patient_id", "age", "gender").show(10)
```
