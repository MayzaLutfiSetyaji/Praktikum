```python
#Implementasi Sederhana MapReduce
#Algoritma Word Count

#Dataset sederhana
def create_simple_dataset():
  documents = [
      "Halo nama saya Mayza Lutfi Setyaji",
      "Saya seorang mahasiswa di Universitas Tidar, Magelang",
      "Saya mengambil jurusan Teknologi Informasi",
      "Bahasa pemrograman yang sedang saat ini saya pelajari adalah Java dan Python",
      "Hari ini saya mengerjakan praktikum Big Data"
  ]
  return documents
#Fungsi Map
def map_function(text):
  for word in text.split():
    yield (word, 1)

#Fungsi Reduce
from collections import defaultdict
def reduce_function(pairs):
  result = defaultdict(int)
  for word, count in pairs:
    result[word] += count
  return result

#Fungsi utama MapReduce
def map_reduce(documents):
  print("=====FASE MAPPING=====") #Fase pertama: Mapping
  mapped_data = []

  for i, doc in enumerate(documents):
        print(f"Dokumen {i+1}: {doc}")
        map_result = list(map_function(doc))
        mapped_data.extend(map_result)  # Menggabungkan semua hasil map
        print(f"Hasil Map: {map_result}")
        print()

  print("=====PENGELOMPOKAN DATA=====") #Fase kedua: Shuffling
  grouped_data = {}

  for word, count in mapped_data:
      if word not in grouped_data:
          grouped_data[word] = []
      grouped_data[word].append(count)

  for word, counts in grouped_data.items():
    print(f"'{word}': {counts}")
  print()

  print("=====FASE REDUCE=====") #Fase ketiga: Reducing
  reduce_input = []
  for word, counts in grouped_data.items():
    # Buat pasangan (word, list_count) untuk reduce
      for count in counts:
          reduce_input.append((word, count))

  final_result = reduce_function(reduce_input)

  return dict(final_result)

# Fungsi untuk menampilkan hasil dengan rapi
def display_results(result):
    print("=====HASIL AKHIR WORD COUNT=====")
    print("-" * 40)
    print(f"{'Kata':<20} {'Jumlah':<10}")
    print("-" * 40)

    # Urutkan berdasarkan jumlah (descending) dan kemudian abjad
    sorted_items = sorted(result.items(), key=lambda x: (-x[1], x[0]))

    for word, count in sorted_items:
        print(f"{word:<20} {count:<10}")

    print("-" * 40)
    print(f"Total kata unik: {len(result)}")

#Program utama
if __name__ == "__main__":
    #Ambil dataset
    dataset = create_simple_dataset()

    print("DATASET YANG DIGUNAKAN:")
    print("=" * 50)
    for i, doc in enumerate(dataset, 1):
        print(f"{i}. {doc}")
    print("=" * 50)
    print()

    # Jalankan MapReduce
    word_count_result = map_reduce(dataset)

    # Tampilkan hasil
    display_results(word_count_result)
```

```python
from collections import defaultdict

def map_function(text):
  for word in text.split():
    yield (word.lower(), 1)

def reduce_function(pairs):
    result = defaultdict(int)
    for word, count in pairs:
        result[word] += count
    return result

text = "Halo nama saya Mayza Lutfi Setyaji, saya suka tidur tapi tidak pernah tidur cukup"
map = list(map_function(text))
reduce = reduce_function(map)

for word, count in reduce.items():
    print(f"{word}: {count}")
```
