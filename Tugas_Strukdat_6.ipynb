```python
# Daftar produk disimpan sebagai list of dictionaries
produk = [
    {"nama": "Rimbamass Whey Gainer rasa Coklat", "harga": 800000, "rating": 4.5},
    {"nama": "Provus Creatine Matrix rasa Strawberry", "harga": 500000, "rating": 4.2},
    {"nama": "Evomass Whey Protein rasa Vanilla", "harga": 300000, "rating": 4.0},
    {"nama": "Provus Creatine Original", "harga": 150000, "rating": 4.7},
    {"nama": "Susu Kedelai Kent Protein", "harga": 100000, "rating": 4.3}
]

# Fungsi untuk menampilkan produk
def tampilkan_produk(list_produk, judul):
    print("\n" + judul)
    print("=" * 40)
    for i, p in enumerate(list_produk, 1):
        print(f"{i}. {p['nama']} - Rp{p['harga']:,} (Rating: {p['rating']})")

#Fungsi Quick Sort
def quick_sort(arr, key='harga', ascending=True):
    if len(arr) <= 1:
        return arr

    pivot = arr[len(arr) // 2]

    if ascending:
        left = [x for x in arr if x[key] < pivot[key]]
        middle = [x for x in arr if x[key] == pivot[key]]
        right = [x for x in arr if x[key] > pivot[key]]
    else:
        left = [x for x in arr if x[key] > pivot[key]]
        middle = [x for x in arr if x[key] == pivot[key]]
        right = [x for x in arr if x[key] < pivot[key]]

    return quick_sort(left, key, ascending) + middle + quick_sort(right, key, ascending)
# Menu utama
def main():
    while True:
        print("\n=== Guntur Geni Gym (3G) Shop ===")
        print("Opsi yang tersedia")
        print("1. Tampilkan produk asli")
        print("2. Urutkan harga termurah")
        print("3. Urutkan harga termahal")
        print("4. Urutkan rating tertinggi")
        print("5. Urutkan rating terendah")
        print("0. Keluar")

        pilihan = input("Pilih menu: ")

        if pilihan == "1":
            tampilkan_produk(produk, "Daftar Produk Asli")
        elif pilihan == "2":
            hasil = quick_sort(produk, 'harga', True)
            tampilkan_produk(hasil, "Produk Termurah ke Termahal (Quick Sort)")
        elif pilihan == "3":
            hasil = quick_sort(produk, 'harga', False)
            tampilkan_produk(hasil, "Produk Termahal ke Termurah (Quick Sort)")
        elif pilihan == "4":
            hasil = quick_sort(produk, 'rating', False)
            tampilkan_produk(hasil, "Produk Rating Tertinggi ke Terendah (Quick Sort)")
        elif pilihan == "5":
            hasil = quick_sort(produk, 'rating', True)
            tampilkan_produk(hasil, "Produk Rating Terendah ke Tertinggi (Quick Sort)")
        elif pilihan == "0":
            print("Terima kasih! Program selesai.")
            break
        else:
            print("Pilihan tidak valid!")

if __name__ == "__main__":
    main()
```
