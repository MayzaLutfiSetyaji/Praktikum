```python
# Daftar produk disimpan sebagai list of dictionaries
produk = [
    {"nama": "Rimbamass Whey Gainer rasa Coklat", "harga": 800000, "rating": 4.5},
    {"nama": "Provus Creatine Matrix rasa Strawberry", "harga": 500000, "rating": 4.2},
    {"nama": "Evomass Whey Protein rasa Vanilla", "harga": 300000, "rating": 4.0},
    {"nama": "Provus Creatine Original", "harga": 150000, "rating": 4.7},
    {"nama": "Susu Kedelai Kent Protein", "harga": 100000, "rating": 4.3}
]

# Fungsi untuk menampilkan produk
def tampilkan_produk(list_produk, judul):
    print("\n" + judul)
    print("=" * 40)
    for i, p in enumerate(list_produk, 1):
        print(f"{i}. {p['nama']} - Rp{p['harga']:,} (Rating: {p['rating']})")

# Sorting harga termurah ke termahal
def sort_harga_murah(produk):
    return sorted(produk, key=lambda x: x['harga'])

# Sorting harga termahal ke termurah
def sort_harga_mahal(produk):
    return sorted(produk, key=lambda x: x['harga'], reverse=True)

# Sorting rating tertinggi ke terendah
def sort_rating_tinggi(produk):
    return sorted(produk, key=lambda x: x['rating'], reverse=True)

# Sorting rating terendah ke tertinggi
def sort_rating_rendah(produk):
    return sorted(produk, key=lambda x: x['rating'])

# Menu utama
def main():
    while True:
        print("\n=== Guntur Geni Gym (3G) Shop ===")
        print("Opsi yang tersedia")
        print("1. Tampilkan produk asli")
        print("2. Urutkan harga termurah")
        print("3. Urutkan harga termahal")
        print("4. Urutkan rating tertinggi")
        print("5. Urutkan rating terendah")
        print("0. Keluar")

        pilihan = input("Pilih menu: ")

        if pilihan == "1":
            tampilkan_produk(produk, "Daftar Produk Asli")
        elif pilihan == "2":
            hasil = sort_harga_murah(produk)
            tampilkan_produk(hasil, "Produk Termurah ke Termahal")
        elif pilihan == "3":
            hasil = sort_harga_mahal(produk)
            tampilkan_produk(hasil, "Produk Termahal ke Termurah")
        elif pilihan == "4":
            hasil = sort_rating_tinggi(produk)
            tampilkan_produk(hasil, "Produk Rating Tertinggi ke Terendah")
        elif pilihan == "5":
            hasil = sort_rating_rendah(produk)
            tampilkan_produk(hasil, "Produk Rating Terendah ke Tertinggi")
        elif pilihan == "0":
            print("Terima kasih! Jangan lupa kembali!")
            break
        else:
            print("Pilihan tidak valid!")

if __name__ == "__main__":
    main()
```
