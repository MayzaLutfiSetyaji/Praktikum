```python
#Tugas Strukdat
class pabrik:
  def __init__(self, gudang):
    self.gudang = gudang
    self.stack = []
  def simpan_barang(self, barang):
    if len(self.stack) < (self.gudang):
      self.stack.append(barang)
      print(f"{barang} berhasil disimpan")
    else:
      print("Gudang penuh!")
  def mengambil_barang(self):
    if self.stack:
      barang = self.stack.pop()
      print(f"Barang {barang} telah diambil")
      return barang
    else:
      print(f"Tidak ada {barang} di gudang!")
      return None
  def tampilkan_barang(self):
    if self.stack:
      print("Daftar barang di gudang: ")
      for barang in reversed(self.stack):
        print(f"- {barang}")
    else:
      print("Gudangnya kosong wak")
kapasitas = int(input("Masukkan kapasitas ruang penyimpanan (2 digit terakhir NPM): "))
Gudang = pabrik(kapasitas)

#Program untuk menampilkan opsi
while True:
  print("\nOpsi:")
  print("1. Tambah dan simpan barang")
  print("2. Ambil barang dari gudang")
  print("3. Berikan daftar barang di gudang")
  print("4. Hentikan program")

  opsi = input("Masukkan opsi (1-4): ")
  if opsi == "1":
    barang = input("Tambahkan barang yang ingin disimpan: ")
    Gudang.simpan_barang(barang)
  elif opsi == "2":
    Gudang.mengambil_barang()
  elif opsi =="3":
    Gudang.tampilkan_barang()
  elif opsi == "4":
    break
  else:
    print(f"Opsi {opsi} tidak tersedia!")

print("Program Berakhir!!")
```
